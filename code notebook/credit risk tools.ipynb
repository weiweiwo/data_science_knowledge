## 非独立事件解释
另一个可能很不可接受的假设是违约独立性的强加。 为什么会有这样的问题呢？ 原因是，尽管信用风险和信用债务人具有高度的特殊性，但它们并不是在真空中运作的。 有很多令人信服的理由表示是依赖违约事件。 特定信贷义务人之间可能存在重要的行业联系。 考虑例如供应链关系或竞争。因此，不良的业绩可能会导致其他可能的信贷对手恶化或改善，又如 一个反欺诈团队有很多笔债务借据，每笔借据其实是有关联的。

在金融业中，负债肯定是相互依存的，这是将某些金融机构视为太大而不能倒闭的正当理由。 我们还应该期待区域或国家的共同风险敞口，以及最普遍意义上的全球经济条件的共同敞口。 如果特定地区，国家或整个星球面临艰难的宏观经济状况，我们应该期望观察到信誉下降的相关性，并最终导致违约。 简而言之，不乏理由不喜欢违约独立性假设。

最好的原因也许源于保守主义原则。 正如在下一章中将非常清楚地看到的那样，依赖关系是信用风险建模的核心。 即使施加相当少量的违约依赖，也会对我们的风险估计产生巨大影响； 可以公平地说，依赖关系是信用风险建模最重要的一个方面。 因此，所有现实和合理的信用风险模型都将依赖结构置于其模型的中心。 话虽如此，仍然可以从简单的独立违约模型中汲取很多教训。 该模型为文献中存在的许多现存的违约风险模型奠定了基础。

```python
# 数值计算风险措施
def computeRiskMeasures(M,lossDistribution,alpha):
    expectedLoss = np.mean(lossDistribution) # 期望损失
    unExpectedLoss = np.std(lossDistribution) # 损失标准差
    expectedShortfall = np.zeros([len(alpha)]) # 预期短缺（也就是右尾之后的平均值）
    var = np.zeros([len(alpha)]) # 风险价值也就百分位时的金额
    for n in range(0,len(alpha)):       # np.ceil  计算大于等于改值的最小整数
        expectedShortfall[n] = np.mean(lossDistribution[np.ceil(alpha[n]*(M-1)).astype(int):M-1]) #         
        var[n] = lossDistribution[np.ceil(alpha[n]*(M-1)).astype(int)] # 因为索引是从0开始，所以10个元素的最大索引为9
    return expectedLoss,unExpectedLoss,var,expectedShortfall   #因此在此M-1个代表为最后一元素的索引

# 蒙特卡洛独立违约
# 思想（1）每次独立随机选择每个交易对象 是否违约
# 操作（2）利用均匀分布在0-1随机生成一个数
# 修改（3）实际逾期概率p如果小于 随机生成的数 就为违约
# 每一次迭代 随机确认违约人数的排序
def independentBinomialLossDistribution(N,M,p,c,alpha,fullOutput=0):
    U = np.random.uniform(0,1,[M,N]) # 生成均匀分布  的 M * N 矩阵 M为迭表示蒙特卡罗迭代次数 N为100共借人的违约概率
    defaultIndicator = 1*np.less(U,p) #随机分布小于 p 为违约标记 
    lossDistribution = np.sort(np.dot(defaultIndicator,c),axis=None) # M 次迭代的总违约风险敞口
    if fullOutput==0:
        return lossDistribution
    else:
        return lossDistribution,defaultIndicator


def independentPoissonLossDistribution(N,M,p,c,alpha):
    lam = -np.log(1-p)
    H = np.random.poisson(lam,[M,N])
    defaultIndicator = 1*np.greater_equal(H,1)
    lossDistribution = np.sort(np.dot(defaultIndicator,c),axis=None)
    return lossDistribution


# 蒙特卡罗二项式分布模型   数值仿真
def independentBinomialSimulation(N,M,p,c,alpha):
    lossDistribution = independentBinomialLossDistribution(N,M,p,c,alpha)
    el,ul,var,es = util.computeRiskMeasures(M,lossDistribution,alpha)
    return el,ul,var,es

# 泊松分布分布模型 对比
def independentPoissonSimulation(N,M,p,c,alpha):
    lossDistribution = independentPoissonLossDistribution(N,M,p,c,alpha)
    el,ul,var,es = util.computeRiskMeasures(M,lossDistribution,alpha)
    return el,ul,var,es
```

```python
# 解析二项式实现
# 
def independentBinomialAnalytic(N,p,c,alpha):
    pmfBinomial = np.zeros(N+1) # 概率质量函数（加入100个人，违约人数情况有 0 到100人，所以在此 N+1）
    for k in range(0,N+1):
        pmfBinomial[k] =  getBC(N,k)*(p**k)*((1-p)**(N-k))  # 公式2.9
    cdfBinomial = np.cumsum(pmfBinomial) # 公式2.35
    varAnalytic = c*np.interp(alpha,cdfBinomial,np.linspace(0,N,N+1)) # 风险价值  # 公式2.30
    esAnalytic = util.analyticExpectedShortfall(N,alpha,pmfBinomial,c) # 预期损失 预期短缺 损失期望值 期望损失
    return pmfBinomial,cdfBinomial,varAnalytic,esAnalytic

# 解析模型中近似的预期差额
def analyticExpectedShortfall(N,alpha,pmf,c):
    cdf = np.cumsum(pmf) # 公式2.35 累计分布函数
    numberDefaults = np.linspace(0,N,N+1)  # 违约数量
    expectedShortfall = np.zeros(len(alpha)) # 预期短缺
    for n in range(0,len(alpha)):    
        myAlpha = np.linspace(alpha[n],1,1000)  # 生成分位数之后（也就是到概率1）的多个x点数（也就是多个分位数）
        nanCheck =  ~np.isnan(pmf) # 空值检查,只提取cdf（x轴）与 逾期人数（y轴）不为空的数

# 注意此处有修改 loss不用拿去做，逾期人数的pmf的线性插值        
#         NumberCount=np.interp(myAlpha,cdf[nanCheck],numberDefaults[nanCheck]) # 修改
        loss = c*np.interp(myAlpha,cdf[nanCheck],numberDefaults[nanCheck]) # 公式2.30  根据线性插值求出每个myAlpha的人数再*c
        prob = np.interp(loss,numberDefaults[nanCheck],pmf[nanCheck])  # 公式2.32 根据线性插值求出每个loss的概率
#         prob = np.interp(NumberCount,numberDefaults[nanCheck],pmf[nanCheck])  # 修改
        
        expectedShortfall[n] = np.dot(loss,prob)/np.sum(prob) # 期望短缺 后面的所有分位数概率乘loss求和 再除以（后的总概率）
    return expectedShortfall
def getBC(N,k): # 等价于scipy.special.comb(N,k) 组合数 不需要排序  # 公式2.9
    a = scipy.special.factorial(N) # N的阶乘
    b = scipy.special.factorial(N-k)  # 阶乘
    c = scipy.special.factorial(k) # 阶乘
    return a/(b*c)  # 组合数公式
```
